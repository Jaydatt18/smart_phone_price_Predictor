# Ploting a top 10 brands of indian market

```python
# brand name
# plot a graph top 10 brand in indian market
sns.set_theme('notebook')
temp_df = df['brand_name'].value_counts().head(10).reset_index()
temp_df
plt.figure(figsize=(7 , 7))
sns.barplot(data = temp_df , x = temp_df['brand_name'], y = temp_df['count'] , palette="Set2")
plt.xticks(rotation='vertical')
plt.show()
```
